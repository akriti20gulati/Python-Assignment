```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

Class: A class is a blueprint for creating objects. It defines a set of attributes and methods that the created objects can use.

Object: An object is an instance of a class. It represents a specific example of the class, with actual values and functionalities defined in the class.

Example:

    class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        print(f"Car brand: {self.brand}, model: {self.model}")

my_car = Car("Toyota", "Corolla")
my_car.display_info()  
```

```python
Q3. Explain why the __init__() function is used. Give a suitable example.

The __init__() function is used in Python to initialize an objects state. It is called when an object is created from a class, allowing the class to initialize the objects attributes with the given values.

Example:
    
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

person1 = Person("Alice", 30)
person1.display()  
```

```python
Q4. Why self is used in OOPs?

The self keyword in Python is used to represent the instance of the class. It allows access to the attributes and methods of the class in Python. self is used as the first parameter in instance methods to refer to the object that is invoking the method.
```

```python
Q5. What is inheritance? Give an example for each type of inheritance.

Inheritance: Inheritance is a feature of OOP that allows a class (child class) to inherit attributes and methods from another class (parent class).

Types of Inheritance:

1. Single Inheritance:
   
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

dog = Dog()
dog.sound()  


2. Multiple Inheritance:
   
class Father:
    def height(self):
        print("Tall")

class Mother:
    def intelligence(self):
        print("High")

class Child(Father, Mother):
    pass

child = Child()
child.height()       
child.intelligence() 

3. Multilevel Inheritance:
   
class Grandparent:
    def heritage(self):
        print("Family Heritage")

class Parent(Grandparent):
    def wisdom(self):
           print("Parental Wisdom")

class Child(Parent):
    pass

child = Child()
child.heritage()
child.wisdom()  
   
4. Hierarchical Inheritance:
   
class Parent:
    def guide(self):
        print("Parent Guidance")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

child1 = Child1()
child2 = Child2()
child1.guide() 
child2.guide() 

5. Hybrid Inheritance: (combination of two or more types of inheritance)
   
class Vehicle:
    def info(self):
        print("This is a vehicle")

class Car(Vehicle):
    def car_info(self):
        print("This is a car")

class Truck(Vehicle):
    def truck_info(self):
        print("This is a truck")

class PickupTruck(Car, Truck):
    def pickup_info(self):
        print("This is a pickup truck")

pickup = PickupTruck()
pickup.info()        
pickup.car_info()    
pickup.truck_info()  
pickup.pickup_info() 
```
